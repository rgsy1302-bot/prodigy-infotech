# Task
Create a histogram to visualize the distribution of the total population in the year 2023 using the data from "/content/API_SP.POP.TOTL_DS2_en_csv_v2_1088695.csv".

## Select data for visualization

### Subtask:
Choose the population data for the year 2023 from the `api_df`.


**Reasoning**:
Select the column for the year 2023 from the api_df DataFrame and store it in a new variable.



```python
population_2023 = api_df['2023']
display(population_2023.head())
```

## Handle missing values

### Subtask:
Address any missing population values for the selected year.


**Reasoning**:
Check for and handle missing values in the '2023' population data, then verify the removal of missing values.



```python
# Check for missing values
missing_before = population_2023.isnull().sum()
print(f"Missing values in population_2023 before dropping: {missing_before}")

# Drop rows with missing values
population_2023 = population_2023.dropna()

# Verify that there are no missing values left
missing_after = population_2023.isnull().sum()
print(f"Missing values in population_2023 after dropping: {missing_after}")
```

## Create histogram

### Subtask:
Generate a histogram to visualize the distribution of the population data for the year 2023.


**Reasoning**:
Generate a histogram to visualize the distribution of the population data for the year 2023 using the cleaned data.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(population_2023, bins=50, edgecolor='black')
plt.title('Distribution of Total Population in 2023')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
