# Task
Analyze and visualize sentiment patterns in social media data from "/content/twitter_training.csv" to understand public opinion and attitudes towards specific topics or brands.

## Load data

### Subtask:
Load the social media data into a dataframe.


**Reasoning**:
Load the social media data into a pandas DataFrame and display the first few rows and information.



```python
import pandas as pd

df = pd.read_csv('/content/twitter_training.csv', header=None)
df.columns = ['id', 'topic', 'sentiment', 'text']
display(df.head())
display(df.info())
```

## Explore data

### Subtask:
Get a sense of the data, including the distribution of sentiment and topics.


**Reasoning**:
Count the occurrences of each unique value in the 'sentiment' and 'topic' columns and display the results to understand their distributions.



```python
sentiment_counts = df['sentiment'].value_counts()
topic_counts = df['topic'].value_counts()

print("Sentiment Distribution:")
display(sentiment_counts)

print("\nTopic Distribution:")
display(topic_counts)
```

## Visualize sentiment distribution

### Subtask:
Create visualizations to show the distribution of sentiment across different topics or brands.


**Reasoning**:
Create a bar plot for sentiment distribution and a bar plot for topic distribution.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values)
plt.title('Distribution of Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(12, 8))
sns.barplot(x=topic_counts.index, y=topic_counts.values)
plt.title('Distribution of Topics')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Reasoning**:
Create a grouped bar plot or a stacked bar plot to visualize the sentiment distribution for each topic.



```python
sentiment_by_topic = df.groupby('topic')['sentiment'].value_counts().unstack().fillna(0)

plt.figure(figsize=(15, 10))
sentiment_by_topic.plot(kind='bar', stacked=True, figsize=(15, 10))
plt.title('Sentiment Distribution by Topic')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```
